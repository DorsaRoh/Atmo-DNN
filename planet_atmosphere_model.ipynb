Model Architecture (deep learning model from scratch)

```python

def initialize_weights(input_dim, hidden_layers, output_dim):
    np.random.seed(42)
    layers = [input_dim] + hidden_layers + [output_dim]
    weights = []
    biases = []
    gamma = []
    beta = []
    for i in range(len(layers)-1):
        weights.append(np.random.randn(layers[i], layers[i+1]) * 0.01)
        biases.append(np.zeros(layers[i+1]))
        gamma.append(np.ones(layers[i+1]))
        beta.append(np.zeros(layers[i+1]))
    return weights, biases, gamma, beta

def relu(Z):
    return np.maximum(0, Z)

def relu_backward(dA, Z):
    dZ = np.array(dA, copy=True)
    dZ[Z <= 0] = 0
    return dZ

def batchnorm_forward(Z, gamma, beta):
    mu = np.mean(Z, axis=0)
    var = np.var(Z, axis=0)
    Z_normalized = (Z - mu) / np.sqrt(var + 1e-8)
    out = gamma * Z_normalized + beta
    cache = (Z, mu, var, gamma, beta)
    return out, cache


def forward_pass(X, weights, biases, gamma=None, beta=None, dropout_ratio=0.5):
    A = X
    cache = {"A0": X}
    drop_cache = {}
    for i in range(len(weights)):
        Z = np.dot(A, weights[i]) + biases[i]
        if i != len(weights) - 1:
            if gamma and beta:  # Check if gamma and beta are provided
                Z, bn_cache = batchnorm_forward(Z, gamma[i], beta[i])
                cache["bn" + str(i+1)] = bn_cache
            A = relu(Z)
            # Dropout
            D = np.random.rand(*A.shape) < dropout_ratio
            A *= D
            A /= dropout_ratio
            drop_cache["D" + str(i+1)] = D
        else:
            A = Z
        cache["Z" + str(i + 1)] = Z
        cache["A" + str(i + 1)] = A
    return A, cache, drop_cache


def update_parameters(weights, biases, grads, learning_rate):
    for i, W in enumerate(weights):
        W -= learning_rate * grads["dW" + str(i + 1)]
        biases[i] -= learning_rate * grads["db" + str(i + 1)]

def compute_cost_with_regularization(Y_hat, Y, weights, lambd):
    m = Y.shape[0]
    L2_regularization_cost = sum(np.sum(np.square(w)) for w in weights)
    return MSE(Y_hat, Y) + (lambd / (2 * m)) * L2_regularization_cost

def backward_pass_with_regularization(Y, cache, weights, lambd):
    m = Y.shape[0]
    dA = cache["A" + str(len(weights))] - Y
    grads = {}
    for i in range(len(weights), 0, -1):
        dZ = relu_backward(dA, cache["Z" + str(i)])
        grads["dW" + str(i)] = (1/m) * np.dot(cache["A" + str(i - 1)].T, dZ) + (lambd/m) * weights[i-1]
        grads["db" + str(i)] = (1/m) * np.sum(dZ, axis=0)
        dA = np.dot(dZ, weights[i-1].T)
    return grads

def get_learning_rate(base_lr, iteration, decay_rate, decay_step):
    return base_lr * (1 / (1 + decay_rate * (iteration/decay_step)))

def train(X, Y, base_learning_rate, decay_rate, num_iterations, hidden_layers, lambd, dropout_ratio=0.5):
    input_dim = X.shape[1]
    output_dim = 1 if len(Y.shape) == 1 else Y.shape[1]
    weights, biases, gamma, beta = initialize_weights(input_dim, hidden_layers, output_dim)
    for i in range(num_iterations):
        learning_rate = get_learning_rate(base_learning_rate, i, decay_rate, 100)
        Y_hat, cache, drop_cache = forward_pass(X, weights, biases, gamma, beta, dropout_ratio)
        cost = compute_cost_with_regularization(Y_hat, Y, weights, lambd)
        grads = backward_pass_with_regularization(Y, cache, weights, lambd)
        update_parameters(weights, biases, grads, learning_rate)
        if i % 100 == 0:
            print(f"Iteration {i}, Cost: {cost}")
    return weights, biases

def MSE(Y_pred, Y_true):
    return np.mean(np.square(Y_pred - Y_true))

def predict(X, weights, biases):
    Y_hat, _, _ = forward_pass(X, weights, biases, dropout_ratio=1)  # Skip gamma and beta here
    return Y_hat

# Call the train function
lambd = 0.7
base_learning_rate = 0.01
decay_rate = 0.01
dropout_ratio = 0.5

# Assuming X_train and Y_train are defined
weights, biases = train(X_train, Y_train, base_learning_rate, decay_rate, num_iterations=1000, hidden_layers=[128, 64, 32], lambd=lambd, dropout_ratio=dropout_ratio)

# Assuming X_test and Y_test are defined
Y_pred = predict(X_test, weights, biases)
print(f"Test MSE: {MSE(Y_pred, Y_test)}")
```

Model Accuracy

```python
def regression_accuracy(Y_pred, Y_true, threshold=0.7):
    """Calculate 'accuracy' for regression.
    
    Args:
    - Y_pred (np.array): Predicted values.
    - Y_true (np.array): True values.
    - threshold (float): The acceptable range for a prediction to be considered 'correct'.
    
    Returns:
    - accuracy (float): Pseudo accuracy for the regression task.
    """
    correct_predictions = np.abs(Y_pred - Y_true) <= threshold
    accuracy = np.mean(correct_predictions) * 100  # in percentage
    return accuracy

# Assuming Y_pred and Y_test are your prediction and ground truth respectively
accuracy = regression_accuracy(Y_pred, Y_test.values)
print(f"Accuracy: {accuracy:.2f}%")
```
