```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
# load dataset
df = pd.read_csv('planetary_data.csv')
df = df.dropna()

# output (targets)
output_columns = ['H20 abundance', 'CO2 abundance',
       'O2 abundance', 'N2 abundance', 'CH4 abundance', 'N2O abundance',
       'CO abundance', 'O3 abundance', 'SO2 abundance', 'NH3 abundance',
       'C2H6 abundance', 'NO2 abundance']

# inputs
features = df.drop(columns=output_columns)
print(features.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

# scale each feature between (0,1)
scaler = MinMaxScaler()
features_scaled = pd.DataFrame(scaler.fit_transform(features),columns=features.columns)
features_scaled
```

Splitting the data into training sets and testing/validation sets!

```python
from sklearn.model_selection import train_test_split

# 40% of data for testing, 60% for training
X_train, X_test = train_test_split(features_scaled, test_size=0.4, random_state=42)
```
